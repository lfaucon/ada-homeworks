## Cleaning

As our first task is to build a skin color predictor, we drop the players for which we do not know the skin color

```python
def skinColor(row):
    return int(4*(row.rater1+row.rater2))

df_clean = df.copy()
df_clean = df_clean[df_clean['rater1'].notnull()]
df_clean = df_clean[df_clean['rater2'].notnull()]
df_clean['skinColor'] = df_clean.apply(skinColor,axis=1)
print(df_clean.groupby('playerShort').first()['skinColor'].value_counts())
print('median', df_clean.groupby('playerShort').first()['skinColor'].median())
df_clean[['rater1','rater2','skinColor']].head()
```

## Aggregate dataset for ML

```python
def getDataset(features):
    df_grouped = df_clean.groupby('playerShort')
    df_grouped_first = df_grouped.first()
    df_players = df_grouped_first.copy()[[]]
    
    for feature in features:
        # This features are continuous but can have NaN values, which we set equal to 0
        if feature in ['height','weight']:
            df_players[feature]= df_grouped_first[feature].apply(lambda x : x if not np.isnan(x) else 0.)
        # These features are categorical so we change them into dummy variables
        if feature in ['club', 'leagueCountry', 'position']:
            df_players = df_players.merge(pd.get_dummies(df_grouped_first[feature]),left_index=True,right_index=True)
        # These features are continuous but not normalized 
        if feature in ['victories','ties','defeats','goals','yellowCards','yellowReds','redCards']:
            df_players['games'] = df_grouped.agg({ 'games': np.sum })
            df_players[feature] = df_grouped.agg({ feature: np.sum })
            df_players[feature+'_per_game'] = df_players.apply(lambda row: row[feature]/row['games'],axis=1) 
    
    df_players_label = df_grouped_first['skinColor'].apply(lambda x: x < 4.)

    return(df_players.values.tolist(),df_players_label.tolist())
```

## Random Forest

```python
def split(X,Y, proportionTrain):
    permutation = np.random.permutation(len(X))
    n = int(proportionTrain*len(X))
    X_train = np.array(X)[permutation[:n]]
    Y_train = np.array(Y)[permutation[:n]]
    X_test = np.array(X)[permutation[n:]]
    Y_test = np.array(Y)[permutation[n:]]
    return (X_train,Y_train,X_test,Y_test)

def runPrediction(X_train,Y_train,X_test,Y_test,parameters):
    for parameter in parameters:
        print('parameter',parameter)
        clf = rfc(
            n_estimators=parameter
        )
        clf.fit(X_train,Y_train)

        train_score = clf.score(X_train,Y_train)
        test_score = clf.score(X_test,Y_test)
        
        print('train score:\t',train_score)
        print('test score:\t',test_score)
        print()
    
def main(features,parameters,proportionTrain):
    (X,Y) = getDataset(features)
    (X_train,Y_train,X_test,Y_test) = split(X,Y,proportionTrain)
    runPrediction(X_train,Y_train,X_test,Y_test,parameters)
    
```

```python
main(
    ['height','weight','club','victories','ties','defeats','goals','yellowCards','yellowReds','redCards'],
    [1,10,100,1000,2000],
    0.8
)
```
