```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = './hillary-clinton-emails/Emails.csv'
```

```python
df_emails = pd.read_csv(filename)
df_emails.columns
```

```python
# Read the whole text.
text = ''
for index, row in df_emails.iterrows():
    text += str(row.ExtractedBodyText) + str(row.ExtractedSubject)
```

```python
text[:1000]
```

```python
# Generate a word cloud image
wordcloud = WordCloud().generate(text)
```

```python
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# lower max_font_size
wordcloud = WordCloud(max_font_size=50).generate(text)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
